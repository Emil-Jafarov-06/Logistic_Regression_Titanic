```python
def categorize(x):
    if(x < 5):
        return 1
    if(x < 15):
        return 2
    if(x < 20):
        return 3
    if(x < 25):
        return 4
    if(x < 30):
        return 5
    if(x < 35):
        return 6
    if(x < 40):
        return 7
    if(x < 50):
        return 8
    if(x < 60):
        return 9
    if(x >= 60):
        return 10
```

```python
X_train['Age']=X_train['Age'].apply(categorize)
```

```python
X_test['Age'] = X_test['Age'].apply(categorize)
```
